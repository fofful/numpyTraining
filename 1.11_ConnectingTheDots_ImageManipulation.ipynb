### Cropping images

```python
a = coffee_cup[100:,50:,:]
```

```python
plt.imshow(a)
plt.show()
```

```python
a1,a2 = np.split(coffee_cup,2)
```

```python
plt.imshow(a1)
plt.show()
```

```python
plt.imshow(a2)
plt.show()
```

```python
b1,b2 = np.split(coffee_cup,3,axis = 1)
```

```python
cropped_coffee_cup = coffee_cup[:,:500,:]
```

```python
b1,b2 = np.split(cropped_coffee_cup,2,axis = 1)
```

```python
plt.imshow(b1)
plt.show
```

```python
plt.imshow(b2)
plt.show
```

### Joining Images

```python
plt.imshow(np.concatenate((a1,a2)))
plt.show()
```

#### By default concatenate occours along the horizontal axis

```python
plt.imshow(np.concatenate((b1,b2)))
plt.show()
```

np.concatenate is stacking the two parts separated horizontally on the top of one another. To avoid this axis = 1 is to be mentioned

```python
plt.imshow(np.concatenate ((b1,b2), axis = 1))
plt.show()
```
